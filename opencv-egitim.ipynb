```python
import cv2
import numpy as np

```

```python
img=cv2.imread("dog0.jpg") #yükler #,cv2.IMREAD_GRAYSCALE görüntüyü siyah beyaz alır ya da direkt 0 vererekte alabiliriz.
img=cv2.resize(img,(200,200)) #img dosyasına yeni size atar.
cv2.namedWindow("image",cv2.WINDOW_NORMAL) #size'ını istediğimiz gibi ayarlarız
cv2.imshow("image",img) #değişkenin ismini image yaptık.
cv2.imwrite("copy.jpg",img) #kopyasını kaydeder
cv2.waitKey(0) # klavyeden herhangi bir tuşa basana kadar bekletir.

cv2.destroyAllWindows() #açık olan tüm pencerelerin otomatik olarak kapanması
########################################
```

```python
# Her video frame'lerden oluşur.
cap=cv2.VideoCapture(0,cv2.CAP_DSHOW) # 0 yazarsak pc kamerası kabul edilir.
file="C:/Users/Ertuğrul/Desktop/Veri Bilimi ve Makine Öğrenmesi (BTK)/OPENCV/video/1.avi"
codec=cv2.VideoWriter_fourcc("W","M","V","2")
framerate=30
resolution=(640,480)
output=cv2.VideoWriter(file,codec,framerate,resolution)
while True:
    ret,frame=cap.read() #ret True veya false döndürür,video bittiği zaman False döndürücek ve sonlandırma işlemi yapacağız. frame ise anlık fotoğrafları yakalar.
    frame=cv2.flip(frame,1) # sağ sol kamera tersliğini engeller
    output.write(frame)
    cv2.imshow("webcab",frame)
    if cv2.waitKey(1) & 0xFF ==ord("q"):
        break
    cv2.waitKey(30) #burayı 1000 yaparsak saniyede 1 frame yakalamaya çalışır
output.release()
cap.release()
cv2.destroyAllWindows()
```

```python
#şekil işlemleri ve üzerlerine text eklemesi
canvas=np.zeros((500,500,3),np.uint8)+255
"""
cv2.line(canvas,(100,100),(300,300),(0,255,0),thickness=4)
"""
f1=cv2.FONT_HERSHEY_SIMPLEX
f2=cv2.FONT_HERSHEY_COMPLEX
f3=cv2.CAP_ARAVIS

cv2.putText(canvas,"Rectangle",(170,150),f1,1.5,(255,0,0),2,cv2.LINE_AA)
cv2.rectangle(canvas,(200,175),(300,250),(255,0,0),4)

cv2.putText(canvas,"Circle",(40,200),f2,1,(0,255,0),1,cv2.LINE_AA)
cv2.circle(canvas,(100,100),(60),(0,255,0),4)

u1=(300,300)
u2=(300,400)
u3=(350,350)

cv2.putText(canvas,"Triangle",(275,290),f3,1,(0,0,255),1,cv2.LINE_AA)
cv2.line(canvas,u1,u2,(0,0,255),4)
cv2.line(canvas,u2,u3,(0,0,255),4)
cv2.line(canvas,u1,u3,(0,0,255),4)

cv2.imshow("image",canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
