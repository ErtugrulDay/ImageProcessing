```python
#resim birleştirme

circle=np.zeros((512,512,3),np.uint8)+255
cv2.circle(circle,(256,256),50,(0,255,0),-1)

rectangle=np.zeros((512,512,3),np.uint8)+255
cv2.rectangle(rectangle,(150,150),(350,350),(0,0,255),-1)

cv2.imshow("circle",circle)
cv2.imshow("rectangle",rectangle)

# A n B
sum=cv2.add(rectangle,circle)
cv2.imshow("A n B",sum)

# A U B
weight=cv2.addWeighted(circle,0.5,rectangle,0.5,0)
cv2.imshow("A U B",weight)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cap=cv2.VideoCapture(0)
def fonk(x):
    pass
cv2.namedWindow("bar")
cv2.resizeWindow("bar",600,600)

cv2.createTrackbar("alt-h","bar",0,180,fonk)
cv2.createTrackbar("alt-s","bar",0,255,fonk)
cv2.createTrackbar("alt-v","bar",0,255,fonk)

cv2.createTrackbar("ust-h","bar",0,180,fonk)
cv2.createTrackbar("ust-s","bar",0,255,fonk)
cv2.createTrackbar("ust-v","bar",0,255,fonk)

cv2.setTrackbarPos("ust-h","bar",150)
cv2.setTrackbarPos("ust-s","bar",200)
cv2.setTrackbarPos("ust-v","bar",200)


while True:
    ret,frame=cap.read()
    frame=cv2.flip(frame,1)
    frame_hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)

    alt_h=cv2.getTrackbarPos("alt-h","bar")
    alt_s=cv2.getTrackbarPos("alt-s","bar")
    alt_v=cv2.getTrackbarPos("alt-v","bar")

    ust_h=cv2.getTrackbarPos("ust-h","bar")
    ust_s=cv2.getTrackbarPos("ust-s","bar")
    ust_v=cv2.getTrackbarPos("ust-v","bar")

    alt_renk=np.array([alt_h,alt_s,alt_v])
    ust_renk=np.array([ust_h,ust_s,ust_v])

    mask=cv2.inRange(frame_hsv,alt_renk,ust_renk)

    cv2.imshow("m",mask)
    cv2.imshow("f",frame)
    if cv2.waitKey(20) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#filtreleme işlemleri
img1=cv2.imread("car.jpg")

blr=cv2.blur(img1,(15,15))
gb=cv2.GaussianBlur(img1,(15,15),cv2.BORDER_DEFAULT)
mb=cv2.medianBlur(img1,9)
b=cv2.bilateralFilter(img1,9,75,75)

cv2.imshow("orj",img1)
cv2.imshow("gb",gb)
cv2.imshow("mb",mb)
cv2.imshow("b",b)
cv2.imshow("blr",blr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1=cv2.imread("1bit1.png")
img2=cv2.imread("2bit2.png")
bit_and=cv2.bitwise_and(img1,img2)
bit_or=cv2.bitwise_or(img1,img2)
bit_xor=cv2.bitwise_xor(img1,img2)
bit_not1=cv2.bitwise_not(img1)
bit_not2=cv2.bitwise_not(img2)
cv2.imshow("1",img1)
cv2.imshow("2",img2)
cv2.imshow("and",bit_and)
cv2.imshow("not1",bit_not1)
cv2.imshow("not2",bit_not2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# köşe tespiti
import numpy as np
img=cv2.imread("ucgen.png")
gri=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gri=np.float32(gri)
corner=cv2.goodFeaturesToTrack(gri,3,0.01,10)

corner=np.int0(corner) #7.3==> 7, 7.9==> 7 yapar
for c in corner:
    x,y=c.ravel() #düzleştirme işlemi
    cv2.circle(img,(x,y),3,(255,0,0),-1)

import numpy as np
img=cv2.imread("opencv.png")
gri=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gri=np.float32(gri)
corner=cv2.goodFeaturesToTrack(gri,20,0.01,10)

corner=np.int0(corner) 
for c in corner:
    x,y=c.ravel() 
    cv2.circle(img,(x,y),3,(255,0,0),-1)
cv2.imshow("img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# canny kenar tespiti
cap=cv2.VideoCapture(0)

while True:
    ret,frame=cap.read()
    frame=cv2.flip(frame,1)
    kenar=cv2.Canny(frame,100,200) #buradan hassasiyeti ayarlıyoruz.

    cv2.imshow("f",frame)
    cv2.imshow("k",kenar)

    if cv2.waitKey(5) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#contour belirleme
img=cv2.imread("insan-vazo.jpg")
gri=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

ret,thresh=cv2.threshold(gri,100,200,cv2.THRESH_BINARY)
cont,a=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE) #diğer 2 parametre default
cv2.drawContours(img,cont,-1,(0,255,0),2) #3.parametre -1 olunca sağ sol ikisini de verir. -1<x<1 

cv2.imshow("1",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Ağırlık merkezi bulma
img=cv2.imread("ucgen.png")

gri=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ai,thresh=cv2.threshold(gri,100,200,cv2.THRESH_BINARY)
m=cv2.moments(thresh)
print(m)
x=int(m["m10"]/m["m00"])
y=int(m["m01"]/m["m00"])

cv2.circle(img,(x,y),3,(0,255,0),-1)
cv2.imshow("s",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
