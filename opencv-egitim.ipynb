cvtColor ile BG2GRAY kullanarak işlem yapmamızın sebebi, genellikle işlem yapılacak görüntü için daha uygun ve daha az işlem gücü gerektiren bir renk uzayı olmasıdır.Gri tonlamalı görüntü, yalnızca bir renk kanalı içerir ve gri tonlarla ifade edilir, bu nedenle işlem yaparken daha basit ve hızlı olabilir.

Görüntü işleme uygulamalarında, şekillerin çevrelerini belirlemek ve konturlarını bulmak için genellikle ikili(binary) görüntüler kullanılır.İkili görüntülerde pikseller yalnızca iki değer alır: siyah(0), veya beyaz(1 veya 255).İkili görüntü, genellikle temel şekil analizi ve nesne tespiti için ön işlem adımı olarak kullanılır ve kontur tespiti için oldukça yaygın bir yaklaşımdır. threshold fonksiyonunda iki parametre olarak verilen değerler şunlardır:
* ret, eşiklemede kullanılan eşik değerini döndüren bir değişkendir. Eğer fonksiyon başarılı bir şekilde çalışırsa True döndürür ve ret kullanılarak eşik değeri elde edilebilir.
* thresh, threshold fonksiyonu sonucu elde edilen ikili görüntüdür.

findContours(), ikili bir görüntü üzerindeki nesnelerin konturlarını(sınırlarını) bulmak için kullanılır. ilk parametresi görüntü (ikili görüntü) üzerinde konturları bulur. İkinci parametre kontur bulma yöntemini ve konturun içinde bulunma ilişkisini belirler. 3.parametre kontur yaklaşım yöntemini belirler.
* contours değişkeni, tespit edilen konturların koordinatlarını içeren bir liste olarak döner.
* hierarchy değişkeni ise kontur hiyerarşisini içeren bir dizidir. Bu hiyerarşi, konturların iç içe geçmiş durumlarını ve üst-alt ilişkilerini ifade eder.Bu özel bir yapıdır ve çeşitli amaçlar için kullanılabilir.


```python
#convex hull ile dış bükey tespit
import numpy as np
import cv2


img=cv2.imread("convexhull.png")

gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh=cv2.threshold(gray,75,200,cv2.THRESH_BINARY) #görüntüyü belirli bir eşik değeriyle ikili bir görüntüye dönüştürmek için cv2.threshold
contur,hierarchy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

h=[]

for i in range(len(contur)):
    h.append(cv2.convexHull(contur[i],False))

z=np.zeros((thresh.shape[0],thresh.shape[1],3),np.uint8) #shape[0],shape[1] ikili görüntünün yükseklik ve genişliğini ifade eder.

for i in range(len(contur)):
    cv2.drawContours(z,contur,i,(255,0,0),3,0)
    cv2.drawContours(z,h,i,(0,255,0),1,0)
# for i in range(len(h)):    
#     cv2.drawContours(z,h,i,(0,255,0),1,0)

#çevre ve alanı bulmak istesek,
cnt=h[0]
M=cv2.moments(cnt) 
print(M)

alan=cv2.contourArea(cnt)
print(alan)
print(M["m00"]) #m00 değeri de aynı zamanda alanı verir.

cevre=cv2.arcLength(cnt,True) #Kapalı bir şekilse True, değilse False
print(cevre)

# cv2.imshow("image",z)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
#Şekil algılama
img=cv2.imread("sekiller.png")
font=cv2.FONT_HERSHEY_SIMPLEX
gri=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

a, thresh=cv2.threshold(gri,200,300,cv2.THRESH_BINARY)
kontur,b=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

for i in kontur:
    e=0.01*cv2.arcLength(i,True)
    approx=cv2.approxPolyDP(i,e,True)
    cv2.drawContours(img,[approx],0,5)

    x=approx.ravel()[0]
    y=approx.ravel()[1]
    print(len(approx))

    if len(approx)==3:
        cv2.putText(img,"ucgen",(x,y),font,1,1,(0))
    elif len(approx)==4:
        cv2.putText(img,"dortgen",(x,y),font,1,1,(0))
    elif len(approx)==5:
        cv2.putText(img,"besgen",(x,y),font,1,1,(0))
    elif len(approx)==6:
        cv2.putText(img,"altigen",(x,y),font,1,1,(0))
    else:
        cv2.putText(img,"cember",(x,y),font,1,1,(0))

cv2.imshow("image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Hough Line Transform Resimden çizgi tespiti

img=cv2.imread("cizgi.png")
gri=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
kenar=cv2.Canny(gri,100,250)
cizgi=cv2.HoughLinesP(kenar,1,np.pi/180,50)
print(cizgi)

for i in cizgi:
    x1,y1,x2,y2=i[0]
    cv2.line(img,(x1,y1),(x2,y2),(255,0,0),2)

cv2.imshow("image",img)
cv2.waitKey(0)
```

```python
cap=cv2.VideoCapture("cizgitakip3.mp4")

while True:
    ret,frame=cap.read()
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    alt=np.array([34,68,94],np.uint8)
    ust=np.array([26,80,62],np.uint8)

    mask=cv2.inRange(hsv,ust,alt)
    kenar=cv2.Canny(mask,50,250)
    cizgi=cv2.HoughLinesP(kenar,1,np.pi/180,50,maxLineGap=5)
    if cizgi is not None:
        for i in cizgi:
            x1,y1,x2,y2=i[0]
            cv2.line(frame,(x1,y1),(x2,y2),(0,255,0),5)
    else:
        print("çizgi tespit edilemedi.")
    cv2.imshow("a",mask)
    cv2.imshow("f",frame)

    if ret==0:
        break
    if cv2.waitKey(20) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
