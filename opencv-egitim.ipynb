cvtColor ile BG2GRAY kullanarak işlem yapmamızın sebebi, genellikle işlem yapılacak görüntü için daha uygun ve daha az işlem gücü gerektiren bir renk uzayı olmasıdır.Gri tonlamalı görüntü, yalnızca bir renk kanalı içerir ve gri tonlarla ifade edilir, bu nedenle işlem yaparken daha basit ve hızlı olabilir.

Görüntü işleme uygulamalarında, şekillerin çevrelerini belirlemek ve konturlarını bulmak için genellikle ikili(binary) görüntüler kullanılır.İkili görüntülerde pikseller yalnızca iki değer alır: siyah(0), veya beyaz(1 veya 255).İkili görüntü, genellikle temel şekil analizi ve nesne tespiti için ön işlem adımı olarak kullanılır ve kontur tespiti için oldukça yaygın bir yaklaşımdır. threshold fonksiyonunda iki parametre olarak verilen değerler şunlardır:
* ret, eşiklemede kullanılan eşik değerini döndüren bir değişkendir. Eğer fonksiyon başarılı bir şekilde çalışırsa True döndürür ve ret kullanılarak eşik değeri elde edilebilir.
* thresh, threshold fonksiyonu sonucu elde edilen ikili görüntüdür.

findContours(), ikili bir görüntü üzerindeki nesnelerin konturlarını(sınırlarını) bulmak için kullanılır. ilk parametresi görüntü (ikili görüntü) üzerinde konturları bulur. İkinci parametre kontur bulma yöntemini ve konturun içinde bulunma ilişkisini belirler. 3.parametre kontur yaklaşım yöntemini belirler.
* contours değişkeni, tespit edilen konturların koordinatlarını içeren bir liste olarak döner.
* hierarchy değişkeni ise kontur hiyerarşisini içeren bir dizidir. Bu hiyerarşi, konturların iç içe geçmiş durumlarını ve üst-alt ilişkilerini ifade eder.Bu özel bir yapıdır ve çeşitli amaçlar için kullanılabilir.


```python
#convex hull ile dış bükey tespit
import numpy as np
import cv2


img=cv2.imread("convexhull.png")

gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh=cv2.threshold(gray,75,200,cv2.THRESH_BINARY) #görüntüyü belirli bir eşik değeriyle ikili bir görüntüye dönüştürmek için cv2.threshold
contur,hierarchy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

h=[]

for i in range(len(contur)):
    h.append(cv2.convexHull(contur[i],False))

z=np.zeros((thresh.shape[0],thresh.shape[1],3),np.uint8) #shape[0],shape[1] ikili görüntünün yükseklik ve genişliğini ifade eder.

for i in range(len(contur)):
    cv2.drawContours(z,contur,i,(255,0,0),3,0)
    cv2.drawContours(z,h,i,(0,255,0),1,0)
# for i in range(len(h)):    
#     cv2.drawContours(z,h,i,(0,255,0),1,0)

#çevre ve alanı bulmak istesek,
cnt=h[0]
M=cv2.moments(cnt) 
print(M)

alan=cv2.contourArea(cnt)
print(alan)
print(M["m00"]) #m00 değeri de aynı zamanda alanı verir.

cevre=cv2.arcLength(cnt,True) #Kapalı bir şekilse True, değilse False
print(cevre)

# cv2.imshow("image",z)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
#Şekil algılama
img=cv2.imread("sekiller.png")
font=cv2.FONT_HERSHEY_SIMPLEX
gri=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

a, thresh=cv2.threshold(gri,200,300,cv2.THRESH_BINARY)
kontur,b=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

for i in kontur:
    e=0.01*cv2.arcLength(i,True)
    approx=cv2.approxPolyDP(i,e,True)
    cv2.drawContours(img,[approx],0,5)

    x=approx.ravel()[0]
    y=approx.ravel()[1]
    print(len(approx))

    if len(approx)==3:
        cv2.putText(img,"ucgen",(x,y),font,1,1,(0))
    elif len(approx)==4:
        cv2.putText(img,"dortgen",(x,y),font,1,1,(0))
    elif len(approx)==5:
        cv2.putText(img,"besgen",(x,y),font,1,1,(0))
    elif len(approx)==6:
        cv2.putText(img,"altigen",(x,y),font,1,1,(0))
    else:
        cv2.putText(img,"cember",(x,y),font,1,1,(0))

cv2.imshow("image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
cap=cv2.VideoCapture("cizgitakip3.mp4")

while True:
    ret,frame=cap.read()
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    alt=np.array([34,68,94],np.uint8)
    ust=np.array([26,80,62],np.uint8) #alt üst ayarlar için color picker sitesinden tespite uygun renk için sınırlar belirleme

    mask=cv2.inRange(hsv,ust,alt)
    kenar=cv2.Canny(mask,50,250)
    cizgi=cv2.HoughLinesP(kenar,1,np.pi/180,50,maxLineGap=5)
    if cizgi is not None:
        for i in cizgi:
            x1,y1,x2,y2=i[0]
            cv2.line(frame,(x1,y1),(x2,y2),(0,255,0),5)
    else:
        pass# print("çizgi tespit edilemedi.")
    cv2.imshow("a",mask)
    cv2.imshow("f",frame)

    if ret==0:
        break
    if cv2.waitKey(20) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#Renk ile nesne tespiti

import cv2
import numpy as np
from collections import deque
#Buffer size ve pts değişkenleri, hareketin izlenmesi için bir buffer mekanizması oluşturmak için kullanılır.
buffer_size=16
pts=deque(maxlen=buffer_size) #koleksiyon veri yapısıdır ve hem baş hem de sondan veri ekleme veya çıkarma işlemleri yapılır.
#eğer deque'nün boyutu "buffer_size'a ulaşırsa, en eski veri otomatik olarak deque'nin başından çıkarılır ve yeni veri deque'nin sonuna eklenir,bu sayede boyut sabit kalır."
bluelower=(84,98,0)
blueupper=(170,200,200)

cap=cv2.VideoCapture(0)
cap.set(3,960) #genişliği 960 piksel olarak ayarlar.
cap.set(4,480) # yüksekliği 480 piksel olarak ayarlar.

while True:
    success,imgorginal=cap.read()
    imgorginal=cv2.flip(imgorginal,1)

    if success:
        blurred=cv2.GaussianBlur(imgorginal,(11,11),0) #görüntüyü yumuşatır ve gürültüyü azaltır
        hsv=cv2.cvtColor(blurred,cv2.COLOR_BGR2HSV)
        cv2.imshow("hsv penceresi",hsv)
        
        #inrange ile elde ettiğimiz maske, yüksekliği ve genişliği değiştirilebilen, genellikle düzensiz ve küçük beyaz bölgelerle dolu olabilir.
        #Erode, bu küçük beyaz bölgeleri küçültüp kaldırarak maskenin daha temiz bir versiyonunu elde etmemizi sağlar.Erozyon sonucunda renk aralığına
        #uyan bölgeler küçülür ve kenarları düzgünleşir.

        #Daha sonra genişletme işlemi(dilate) uygulanır.Bu, beyaz bölgeyi tekrar genişletir ve kenarları daha düzgün hale getirir.Erozyon ve genişletme işlemleri,
        #birbirini takip eden işlemler olarak kullanılır ve bu işlem sırası, maskenin daha tutarlı ve dengeli olmasını sağlar.
        mask=cv2.inRange(hsv,bluelower,blueupper) #bu renk aralığında kalan pikseller beyaz,diğerleri siyah olacak şekilde maske oluşur.
        mask=cv2.erode(mask,None,iterations=2)
        mask=cv2.dilate(mask,None,iterations=2)
        cv2.imshow("maskeli pencere",mask)


        
        
        """Maskeleme yapilan görüntünün konturlarinin bulunmasi,bu konturlardan en büyük olaninin belirlenmesi,minimum alan dikdörtgeni(bounding box) ve
        döndürme acisinin hesaplanmasi, bunlarin ekrana yazilmasi,döndürülmüs dikdörtgenin ve kontur merkezinin çizilmesi,ardindan çizilen merkezlerin arasinda
        bağlanti çizgileri çizilmesini içeren işlemler aşağidakilerdir. 
        """
        (contours,a)=cv2.findContours(mask.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
        #EXTERNAL,yalnızca dış konturları almak için kullanılır, CHA_AP_SIMPLE_, kontur noktalarını sadece başlangıç ve bitiş noktalarıyla temsil etmek için kullanılır.
        
        center=None #merkez değişkeni
        if len(contours)>0:
            c=max(contours,key=cv2.contourArea) #konturların içinden en büyük alanlı konturu seçer
            rect=cv2.minAreaRect(c) #seçilen konturun etrafına en küçük dikdörtgen kutuyu çizer.
            ((x,y),(width,height),rotation)=rect
            s="x:{}, y:{}, width:{}, height:{}".format(np.round(x),np.round(y),np.round(width),np.round(height),np.round(rotation))
            print(s)

            box=cv2.boxPoints(rect) # dikdörtgenin dört köşe noktasını alır.
            box=np.int64(box) #köşe noktalarını integer'a dönüştürür.

            M=cv2.moments(c) #kontor merkez hesaplama ve hemen ardından center değişkenine atama
            center=(int(M["m10"]/M["m00"]),int(M["m01"]/M["m00"]))
            cv2.drawContours(imgorginal,[box],0,(0,255,0),2) #dikdörtgenin kontürünü yeşil renkte çizer
            cv2.circle(imgorginal,center,5,(255,0,0)) #hesaplanan merkez noktasını mavi bir daire olarak çizer

        pts.append(center)
        for i in range(1,len(pts)):
            if pts[i-1] is None or pts[i] is None:continue
            cv2.line(imgorginal,pts[i-1],pts[i],(0,255,255),3)
            #Önceki merkez noktası ile şu anki merkez noktası arasında sarı renkte bir çizgi çizer.Bu, izlenen nesnenin hareketini görselleştirir ve
            # izlenen nesnenin geçmiş konumları arasında bir iz bırakır.
        cv2.imshow("original",imgorginal)

        





    if cv2.waitKey(1) & 0xFF==ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#Open CV ile Maske tespiti

import cv2
cap=cv2.VideoCapture(0)
face_cascade=cv2.CascadeClassifier("frontalface.xml")
mouth_cascade=cv2.CascadeClassifier("mouth.xml")

while True:
    ret,frame=cap.read()
    frame=cv2.flip(frame,1)
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,1.1,7)

    if len(faces)==0:
        cv2.putText(frame,"Yuz tespit edilemedi.",(40,40),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
    else:
        for x,y,w,h in faces:
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
            roi_gray=gray[y:y+h,x:x+w]
            mouth=mouth_cascade.detectMultiScale(roi_gray,1.4,10)
            i=0
            if len(mouth)==0:
                cv2.putText(frame,"Maske var",(40,40),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
            else:
                for x1,y1,w1,h1 in mouth:
                    if i==0:
                        i=i+1
                        cv2.rectangle(frame,(x1+x,y1+y),(x1+x+w1,y1+y+h1),(255,0,0),3)
                    else:
                        pass
                    cv2.putText(frame,"Maske yok",(40,40),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)

    cv2.imshow("a",frame)
    if cv2.waitKey(1) & 0xFF==ord("q"):
        break

cv2.destroyAllWindows()
```

```python
#Trafikte geçen araç sayısını hesaplama with mog
import cv2
import numpy as np
cap=cv2.VideoCapture("trafficvideo2.mp4")
backsub=cv2.createBackgroundSubtractorMOG2(history=100,varThreshold=16,detectShadows=False)#arka planı temizleme işlemi 
c=0
previous_count = 0

width = int(cap.get(3))
height = int(cap.get(4))

print("Video Genişliği:", width)
print("Video Yüksekliği:", height)
while True:
    ret,frame=cap.read()
    if ret==1:
        fgmask=backsub.apply(frame)
        fgmask=cv2.erode(fgmask,None,iterations=2)
        fgmask=cv2.dilate(fgmask,None,iterations=2)

        cv2.line(frame,(0,280),(640,280),(0,255,0),2)
        cv2.line(frame,(0,350),(640,350),(0,255,0),2)


        contours,hiers=cv2.findContours(fgmask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
        try: 
            hiers=hiers[0]
        except: 
            hiers=[]


        for contour,hier in zip(contours,hiers):
            (x,y,w,h)=cv2.boundingRect(contour)
            if w>60 and h>60:
                cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
                if y>280 and y<350:
                    c=c+1
        if c>previous_count:
            previous_count=c

        font=cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(frame,"car "+str(previous_count),(90,100),font,2,(0,0,255),2,cv2.LINE_AA)
    
        cv2.imshow("vid",frame)
        cv2.imshow("vid2",fgmask)
        if cv2.waitKey(35) & 0xFF==ord("q"):
            break
    
    
cap.release()
cv2.destroyAllWindows()
```
